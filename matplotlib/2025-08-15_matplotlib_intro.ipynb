```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [20,25,30,35,40,45,50,55,60]
weight_list = [70,75,73,85,89,90,75,84,90]
```

```python
age_list_numpy = np.array(age_list)
weight_list_numpy = np.array(weight_list)
```

```python
age_list_numpy
```

```python
weight_list_numpy
```

```python
plt.plot(age_list_numpy, weight_list_numpy)
```

```python
plt.plot(age_list_numpy, weight_list_numpy, "g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age - Weight Graph")
plt.show()
```
