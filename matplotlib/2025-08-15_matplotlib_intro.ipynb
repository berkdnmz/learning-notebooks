```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [20,25,30,35,40,45,50,55,60]
weight_list = [70,75,73,85,89,90,75,84,90]
```

```python
age_list_numpy = np.array(age_list)
weight_list_numpy = np.array(weight_list)
```

```python
age_list_numpy
```

```python
weight_list_numpy
```

```python
plt.plot(age_list_numpy, weight_list_numpy)
```

```python
plt.show()
```

```python
plt.plot(age_list_numpy, weight_list_numpy, "g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age - Weight Graph")
plt.show()
```

```python
my_numpy1 = np.linspace(0,20,25)
```

```python
my_numpy1
```

```python
my_numpy2 = my_numpy1 **3
```

```python
my_numpy2
```

```python
plt.plot(my_numpy1, my_numpy2,"g*-")
plt.show()
```

```python
plt.subplot(1,2,1) #1 row, 2 column, 1.graph
plt.plot(my_numpy1, my_numpy2, "r--")
plt.subplot(1,2,2) 
plt.plot(my_numpy2, my_numpy1, "g*")
plt.show()
```

```python
my_figure = plt.figure()
my_axes= my_figure.add_axes([1,1,0.5,0.5])
my_axes.plot(my_numpy1, my_numpy2, "g*-")
my_axes.set_xlabel("X Data")
my_axes.set_ylabel("Y Data")
my_axes.set_title("Graph Title")
plt.show()
```

```python
my_figure2 = plt.figure()
my_axes2= my_figure2.add_axes([1,1,0.5,0.5])
my_axes2.plot(my_numpy1, my_numpy2, "g*-")
my_axes2.set_xlabel("X Data small")
my_axes2.set_ylabel("Y Data small")
my_axes2.set_title("Small Graph")
plt.show()
my_figure3 = plt.figure()
my_axes3= my_figure3.add_axes([0.3,0.3,0.9,0.9])
my_axes3.plot(my_numpy1, my_numpy2, "g*-")
my_axes3.set_xlabel("X Data Large")
my_axes3.set_ylabel("Y Data Large")
my_axes3.set_title("Large Graph")
plt.show()
```

```python
(my_fig, my_axe) = plt.subplots(2,3)
plt.show()
```

```python
(my_fig, my_axe) = plt.subplots()
my_numpy3 = np.linspace(0,15,20)
my_numpy4 = my_numpy3 ** 1.5
my_axe.plot(my_numpy3, my_numpy4, color ="#00FF19", alpha=1)
my_axe.plot(my_numpy4, my_numpy3, color ="#FF00EA", alpha=0.2)
plt.show()
```

```python
(new_fig, new_axe) = plt.subplots()
new_axe.plot(my_numpy3, my_numpy3 + 2, color="#FF0000", linewidth = 3.4)
new_axe.plot(my_numpy3, my_numpy3 + 3, color="#6FFF00", linewidth = 1.4)
new_axe.plot(my_numpy3, my_numpy3 + 4, color="#00FFF2", linewidth = 1.4, linestyle="-.")
new_axe.plot(my_numpy3, my_numpy3 + 5, color="#2F00FF", linewidth = 1.4, linestyle=":")
new_axe.plot(my_numpy3, my_numpy3 + 6, color="#E600FF", linewidth = 1.4, linestyle="--", marker="o")
new_axe.plot(my_numpy3, my_numpy3 + 7, color="#0088FF", linewidth = 1.4, linestyle="--", marker="+", markersize=8)
plt.show()
```
