```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
x = iris.data
y = iris.target
features = iris.feature_names
```

```python
df = pd.DataFrame(x, columns=features)
df["species"] = y
```

```python
df
```

```python
df.info()
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
df
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled,y, test_size=0.3, random_state=60)
```

```python
models = { "Logistic Regresyon": LogisticRegression(),
           "KN": KNeighborsClassifier(),
           "SVM": SVC()}
```

```python
accuracy_values = {}
```

```python
for model_name, model in models.items():
    model.fit(x_train, y_train)
    y_prediction = model.predict(x_test)

    accuracy = accuracy_score(y_test, y_prediction)

    accuracy_values[model_name] = accuracy
```

```python
print("DoÄŸruluk oranlarÄ±")
for model_name, accuracy in accuracy_values.items():
    print(f"{model_name}: {accuracy * 100: .3f}")
```
