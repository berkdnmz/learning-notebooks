```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np

import importnb
with importnb.Notebook():
    from myKmeans import kmeans as mykmeans
with importnb.Notebook():
    from myDBSCAN import dbscan as mydbscan
```

```python
X, y = make_blobs(n_samples=1000, centers=4, cluster_std=2.4, random_state=31)
plt.scatter(X[:,0], X[:,1], c="black", s=100, alpha=1, label="Original Data")
```

```python
clusters, noise = mydbscan(X.tolist(), 1.8, 4)
```

```python
for i, cluster in enumerate(clusters):
    cluster = np.array(cluster)
    plt.scatter(cluster[:,0], cluster[:,1], s=10, label=f"Kme {i}")

# noise'lar siyah x ile iz
if noise:
    noise = np.array(noise)
    plt.scatter(noise[:,0], noise[:,1], c="black", marker="x", s=10, label="Noise")

plt.legend()
plt.show()
```
