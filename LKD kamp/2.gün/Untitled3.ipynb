```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
features = iris.feature_names
df = pd.DataFrame(x, columns= features)
df["species"] = y
```

```python
df
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(121, projection="3d")
ax.scatter(df["sepal length (cm)"], df["sepal width (cm)"], df["petal length (cm)"], 
           c=df["species"], cmap="viridis")
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
pca = PCA(n_components=3)
x_pca = pca.fit_transform(x_scaled)
```

```python
df_pca = pd.DataFrame(x_pca, columns=["PC1", "PC2","PC3"])
df_pca["species"] = y
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(121, projection="3d")
ax.scatter(df_pca["PC1"], df_pca["PC2"], df_pca["PC3"], 
           c=df_pca["species"], cmap="viridis")


'''
plt.figure(figsize=(8,6))
plt.scatter(df_pca["PC1"], df_pca["PC2"], c=df_pca["species"], cmap="viridis")
plt.colorbar(label="Species")
plt.show()'''
```
