```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
path = r"C:/Users/berkd/Desktop/GITHUB/learning-notebooks/LKD kamp/data/verionislem.csv"
df = pd.read_csv(path)
df
```

```python
df.info()
```

```python
# 1. eksik verileri doldurma
```

```python
df["Yaş"].fillna(df["Yaş"].mean(), inplace=True)
df["Maaş"].fillna(df["Maaş"].mean(), inplace=True)
```

```python
df.info()
```

```python
#2 Aykırı değerlerle mücadele
```

```python
Q1 = df["Çalışma Saati"].quantile(0.25)
Q3 = df["Çalışma Saati"].quantile(0.75)
IQR = Q3-Q1
alt_sinir = Q1 - 1.5 * IQR
ust_sinir = Q3 + 1.5 * IQR

(df["Çalışma Saati"] >= alt_sinir) & (df["Çalışma Saati"] >= ust_sinir)
```

```python
df
```

```python
df = df[(df["Çalışma Saati"] >= alt_sinir) & (df["Çalışma Saati"] <= ust_sinir)]
```

```python
df
```

```python
# Kategorik verileri idare etmek
```

```python
label_encoder = LabelEncoder()
df["Şehir"] = label_encoder.fit_transform(df["Şehir"])
df["Departman"] = label_encoder.fit_transform(df["Departman"])
```

```python
df
```

```python
# normalize etmek
```

```python
scaler = MinMaxScaler()
df[["Naaş","Çalışma Saati", "Deneyim Yılı"]] = scaler.fit_transform(df[["Maaş","Çalışma Saati", "Deneyim Yılı"]])
```
